```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Loading the dataset
df = pd.read_csv('Diabetes Classification.csv')
df.head()
```

```python
# Checking the shape of the dataset
df.shape
```

```python
# Checking for null/missing values
df.isnull().sum()
```

```python
# Checking the datatypes
df.dtypes
```

```python
# Unique values in each column
for i in df.columns:
    print(i, df[i].unique())
```

```python
# columns for outlier removal
cols  = ['Age', 'BMI', 'FBS','HbA1c']

#IQR for the selected columns
Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1

#Threshold for outlier removal
threshold = 1.5

#Find index of outliers
index = np.where((df[cols] < (Q1 - threshold * IQR)) | (df[cols] > (Q3 + threshold * IQR)))[0]

#Drop outliers
df = df.drop(df.index[index])
```

```python
from sklearn.preprocessing import LabelEncoder

# List of categorical variables
cols = ['Age', 'Gender', 'BMI', 'Blood Pressure', 'FBS', 'HbA1c','Family History of Diabetes', 'Smoking', 'Diet', 'Exercise','Diagnosis']

# Label encoding object
le = LabelEncoder()

for i in cols:
    le.fit(df[i])
    df[i] = le.transform(df[i])
    print(i, df[i].unique())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns = ['Diagnosis']), df['Diagnosis'], test_size = 0.2, random_state = 0)
```

X_train, X_test, y_train, y_test = train_test_split(...):
This line assigns the resulting train and test sets to the respective variables.

df.drop(columns = ['Diagnosis']):
This part removes the 'Diagnosis' column from the DataFrame df, and the remaining columns are assigned to X.

df['Diagnosis']:
This extracts the 'Diagnosis' column from the DataFrame df, and it is assigned to y.

test_size = 0.2:
This sets the proportion of the dataset to be used for the test set, in this case, 20% of the data.

random_state = 0:
This sets the random state for the split, ensuring that the same train-test split is generated each time the code is run (for reproducibility).

The resulting variables are:

X_train: The feature data for the training set.

X_test: The feature data for the test set.

y_train: The target variable (Diagnosis) for the training set.

y_test: The target variable (Diagnosis) for the test set

## Diabetes Disease Prediction

For predicting the diabetes disease, I have used the following classification models:
1. Random Forest Classifier
2. Decision Tree Classifier

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

# Create Random Forest object
rfc = RandomForestClassifier(random_state=0, max_features='sqrt', n_estimators=200, class_weight='balanced')
```

from sklearn.ensemble import RandomForestClassifier:
This line imports the RandomForestClassifier class from the sklearn.ensemble module.

rfc = RandomForestClassifier(...):
This line creates an instance of the RandomForestClassifier with the specified parameters.

random_state=0:
This sets the random state for the model, ensuring reproducibility of the results.

max_features='sqrt':
This parameter sets the maximum number of features to consider at each split in the decision trees to the square root of the total number of features.

n_estimators=200:
This sets the number of decision trees in the Random Forest to 200.

class_weight='balanced':
This parameter adjusts the class weights to be "balanced", which can help with imbalanced datasets by giving more importance to the minority class.

The RandomForestClassifier is a powerful ensemble learning algorithm that combines multiple decision trees to improve the overall classification performance. It works by training multiple decision trees on random subsets of the training data and features, and then aggregating the predictions from these trees to make the final classification.

```python
# Training the model
rfc.fit(X_train, y_train)
```

rfc.fit(X_train, y_train):
This line trains the Random Forest Classifier model (rfc) on the training data.

X_train represents the feature data for the training set.

y_train represents the target variable (labels) for the training set.

The fit() method is a fundamental part of the machine learning workflow, as it is responsible for training the model on the provided data. During the training process, the Random Forest Classifier algorithm learns the underlying patterns and relationships in the training data, which will enable it to make predictions on new, unseen data.

```python
# Training accuracy
rfc.score(X_train, y_train)*100
```

```python
# Predicting the test set results
rfc_pred = rfc.predict(X_test)
```

 ### Decision Tree Classifier

```python
 from sklearn.tree import DecisionTreeClassifier
# Create Decision Tree object
dtc = DecisionTreeClassifier(random_state=0, max_depth= 12, min_samples_leaf=2, min_samples_split=2, class_weight='balanced')
```

The Decision Tree Classifier is a popular machine learning algorithm that uses a tree-like model of decisions to make predictions. Each internal node in the tree represents a feature, and the branches represent the possible values of that feature. The leaf nodes represent the final classification or prediction.

```python
# Training the model
dtc.fit(X_train, y_train)
```

```python
# Training accuracy
dtc.score(X_train, y_train)*100
```

```python
# Predicting the test set results
dtc_pred = dtc.predict(X_test)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
```

```python
#Training the model
lr.fit(X_train, y_train)
```

```python
#Training accuracy
lr.score(X_train, y_train)*100
```

```python
#Predicting the test set results
lr_pred = lr.predict(X_test)
```

## Model Evalution

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
fig, ax  = plt.subplots(1,3, figsize = (20,5))
sns.heatmap(confusion_matrix(y_test, rfc_pred), annot = True, cmap = 'coolwarm', ax = ax[0]).set_title('Random Forest')
sns.heatmap(confusion_matrix(y_test, dtc_pred), annot = True, cmap = 'coolwarm', ax = ax[1]).set_title('Decision Tree')
sns.heatmap(confusion_matrix(y_test, lr_pred), annot = True, cmap = 'coolwarm', ax = ax[2]).set_title('Logistic Regression')
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('Random Forest')
print('Accuracy Score: ', accuracy_score(y_test, rfc_pred)*100)
print('Precision Score: ', precision_score(y_test, rfc_pred)*100)
print('Recall Score: ', recall_score(y_test, rfc_pred)*100)
print('F1 Score: ', f1_score(y_test, rfc_pred)*100)
```

```python
print('Decision Tree')
print('Accuracy Score: ', accuracy_score(y_test, dtc_pred)*100)
print('Precision Score: ', precision_score(y_test, dtc_pred)*100)
print('Recall Score: ', recall_score(y_test, dtc_pred)*100)
print('F1 Score: ', f1_score(y_test, dtc_pred)*100)
```

```python
print('Logistic Regression')
print('Accuracy Score: ', accuracy_score(y_test, lr_pred)*100)
print('Precision Score: ', precision_score(y_test, lr_pred)*100)
print('Recall Score: ', recall_score(y_test, lr_pred)*100)
print('F1 Score: ', f1_score(y_test, lr_pred)*100)
```
